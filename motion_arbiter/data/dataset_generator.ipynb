```python
import pandas as pd
import csv
import re
from nltk.tokenize import word_tokenize
```

```python
def cleaning(sentences):
    words = []
    for s in sentences:
        clean = re.sub(r'[^ a-z A-Z 0-9]', " ", s)
        w = word_tokenize(clean)
        #stemming
        words.append([i.lower() for i in w])
    
    return words  
```

```python
utterance = []
label = []

with open('restaurant_all.tsv') as tsvfile:
    reader = csv.DictReader(tsvfile, dialect='excel-tab')
    for row in reader:
        utter = row['Message.Text']
        intent = row['Dialog-Acts'].split('(')[0]
        
        if len(utter)>10 and not(utter.startswith('pro')) and not(utter.startswith('0')) and len(utter)<100:
            if len(intent) > 0 and (intent != 'Deny') and (intent != 'confirm_question') and (intent != 'Not_Sure') and (intent != 'not_sure'):
#             if len(intent) > 0:
                utterance.append(row['Message.Text'])
                label.append(row['Dialog-Acts'].split('(')[0])
# set(label)
cleaned = cleaning(utterance)
for u in cleaned:
    utter = ' '.join(u)
    utterance.append(utter)
# print(label[:10])
list(set(label))
```

```python
file = open('intent_data.csv', 'w')
for x,y in zip(utterance,label):
    if y != '':
        x = re.sub(r'[^ a-z A-Z 0-9]', " ", x)
        file.write(x+','+y+'\n')
```
